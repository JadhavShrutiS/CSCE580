```python
#First all the modules that are required are imported.
import fitz
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import nltk
from nltk.corpus import stopwords
from pathlib import Path
import os
```

TASK 1: INDIVIDUAL RESUME ANALYSIS

```python
#Read in resume in text and get a list of words, called resume_words
doc = fitz.open('C:/Shruti/College/Semesters/Fall 2025/CSCE580/Quiz1/Resume_Updated_ShrutiJ.pdf')
resume_words = ""
for page in doc:
   resume_words+=page.get_text()
print(resume_words)
```

a) Plot histogram of top 20 resume_words

```python
#first count top 20 words
counter = Counter(resume_words.split())
commonwords = counter.most_common(20)
words = list(t[0] for t in commonwords)
frequency  = list(t[1] for t in commonwords)

#plot top 20 words
plt.bar(words, frequency)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top words-Self Resume')
plt.xticks(rotation=90)
# Display the plot
plt.show()
```

This graph contains a lot of stop words according to the list provided: https://gist.github.com/sebleier/554280

b) Remove stop_words from resume_words to get specific_words.

```python
#remove stop words to get specific words
stop_words = set(stopwords.words('english'))
words_temp = resume_words.split() 
filtered_words = [word for word in words_temp if word not in stop_words]
specific_words= ' '.join(filtered_words) 

#find top 20 words after cleaning
counter1 = Counter(specific_words.split())
commonwords1 = counter1.most_common(20)
words1 = list(t[0] for t in commonwords1)
frequency1  = list(t[1] for t in commonwords1)

#plot the frequency of new top 20 words
plt.bar(words1, frequency1)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top words after removing stop words- Self Resume')
plt.xticks(rotation=90)
plt.show()
```

c) Analysis: Note which words emerge now.

Question: Was removing stop_words helpful in revealing more about you?

Answer: Yes, removing stop_words was helpful in revealing a little bit more as it revealed words such as showed other important words like CubeSat and Engineering giving a little more insight into my project and background.
